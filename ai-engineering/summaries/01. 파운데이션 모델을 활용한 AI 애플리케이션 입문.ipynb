## 1.1. AI 엔지니어링의 부상
### 1.1.1. 언어 모델에서 대규모 언어 모델로
- 이전에 존재하던 언어 모델이 자기 지도 학습으로 지금과 같은 규모로 성장
- 모델의 크기는 일반적으로 파라미터의 수로 측정
  - 책 집필 시점 기준 1,000억 개의 파라미터 가진 모델을 대규모로 간주

#### 언어 모델
- 하나 이상의 언어에 대한 통계 정보 인코딩
- 기본 단위는 **토큰**
  - 언어 모델에 따라 토큰은 문자, 단어, 단어의 일부가 될 수 있음
- 모델이 다룰 수 있는 모든 토큰의 집합을 모델의 **어휘**라고 부름
- 토큰을 예측할 때 사용할 수 있는 정보에 따라 **마스크 언어 모델**과 **자기회귀 언어 모델**로 나뉨
- 마스크 언어 모델
  - 누락된 토큰 전후 컨텍스트를 사용해 시퀀스 어느 위치에서든 누락된 토큰을 예측하도록 학습
- 자기회귀 언어 모델
  - 이전 토큰들만 보고 시쿼스의 다음 토큰을 예측하도록 학습
- 출력에 제한 없음
  - 정해진 답 없이 개방형 출력을 생성하는 모델을 생성형 모델이라고 부름
  - 생성형 AI라는 용어가 유래

#### 자기 지도 학습
- 지도 학습
  - 레이블이 있는 데이터를 사용해 ML 알고리즘을 학습하는 과정
  - 데이터 레이블링에 많은 비용과 시간 소요
- 자기 지도 학습에서는 명시적인 레이블이 필요하지 않고 모델 입력 데이터에서 레이블 추론

### 1.1.2. 대규모 언어 모델에서 파운데이션 모델로
- AI가 세상에 유의미하게 작동하기 위해서 글자 이상의 데이터 처리하는 능력이 필요해, 더 많은 모달리티 포함하도록 확장
- 둘 이상의 데이터 형태를 처리할 수 있는 모델을 멀티모달 모델이라고 함
- 생성형 멀티모달 모델은 대규모 멀티모달 모델이라고도 함
- 모델들이 AI 애플리케이션에서 갖는 중요성과 다양한 요구사항에 맞게 발전시킬 수 있다는 점에서 **파운데이션 모델**로 설명하는 게 적절
- 특정 작업에 대한 성능을 올리고 싶다면 아래 기법 사용 가능
  - 특정 작업 관련 데이터셋을 사용해 모델을 추가로 **파인튜닝 (fine-tuning)**
  - 상세한 지시를 모델에 제공하는 **프롬프트 엔지니어링 (prompt engineering)**
  - 데이터베이스를 활용해 지시를 보완하는 **검색 증강 생성 (retrieval-augmented generation, RAG)**

### 1.1.3. 파운데이션 모델에서 AI 엔지니어링으로
- **AI 엔지니어링**은 파운데이션 모델을 기반으로 애플리케이션 만드는 과정을 의미
- 전통적인 ML 엔지니어링이 모델 자체를 개발한다면, AI 엔지니어링은 이미 존재하는 모델을 활용
- AI 엔지니어링의 빠른 성장에 기여하는 요인
  - 범용 AI 능력
    - 파운데이션 모델로 더 많은 작업 수행
  - AI 투자 증가
  - AI 애플리케이션 개발에 대한 낮아진 진입 장벽
    - 오픈AI 및 기타 모델 제공업체가 대중화한 서비스형 모델 접근 방식으로 AI를 활용해 애플리케이션 더 쉽게 개발

## 1.3. AI 애플리케이션 기획
### 1.3.1. 활용 사례 평가
- AI 애플리케이션을 왜 만들고 싶은지 질문
- 직접 만들어야 하는지 고려
  - AI가 비즈니스 생존에 결정적인 영향을 미친다면 직접 개발
  - 수익과 생산성을 높이기 위해서라면, 기존 솔루션 구매하는 것일 시간과 비용을 절약하며 더 좋은 성능 얻을 수 있음

#### 애플리케이션에서 AI와 사람의 역할
- AI 제품에서 AI가 맡는 역할에 따라 애플리케이션 개발 방식과 요구사항이 달라짐
  - 핵심적 또는 보완적
    - 앱이 AI 없이도 작동할 수 있다면 AI는 보완적
  - 반응형 또는 선제형
    - **반응형**은 사용자 요청이나 특정 행동에 응답하는 방식으로 작동
      - 이벤트에 즉시 응답해야 해 지연 시간이 짧아야 함
    - **선제형**은 사용자에게 유용한다고 판단되는 적절한 시점에 먼저 정보 제시
      - 지연 시간의 중요성이 상대적으로 덜함
      - 일반적으로 더 높은 품질 기준을 요구
  - 동적 또는 정적
    - **동적 방식**은 사용자 피드백으로 지속적으로 업데이트
      - 각 사용자가 자신의 데이터로 모델 파인튜닝하며 자신만의 모델을 가짐
      - ChatGPT의 메모리 기능과 같은 개인화 장치 둘 수 있음
    - **정적 방식**은 주기적으로만 업데이트
      - 하나의 공유 모델을 여러 사용자가 사용
      - 모델 성능 개선은 공유 모델 업데이트될 때만 일괄적으로 이루어짐
- 애플리케이션에서 사람의 역할 (AI가 사람을 뒤에서 지원할지, 직접 결정 내릴지, 둘 다 할 것인지)을 명확히 하는 것도 중요
  - 휴먼 인 더 루프 (human-in-the-loop)
    - AI의 의사결정 과정에 사람을 참여시키는 것을 의미
  - 크롤-워크-런 (crawl-walk-run)
    - 마이크로소프트에서 AI 자동화 점진적으로 증가시키기 위한 제안한 프레임워크
    - **크롤**은 사람의 참여가 필수 임을 의미
    - **워크**는 AI가 내부 직원과 직접 상호작용할 수 있음을 의미
    - **런**은 잠재적으로 외부 사용자와의 직접적인 AI 상호작용을 포함해 자동화가 향상

#### AI 제품 방어의 가능성
- 진입 장벽이 낮기 때문에 방어 가능성 고려해야 함
- 기술력, 데이터, 유통력이 AI 분야의 경쟁 우위가 됨

### 1.3.2. 기대치 설정
- AI 애플리케이션이 비즈니스에 어떤 영향을 미칠지 파악해야 함
- 제품을 고객에게 공개하기 전 필요한 최소 성능이 어느 정도인지 명확히 정해야 함

### 1.3.3. 마일스톤 계획
- 이미 존재하는 모델의 성능을 평가해 능력을 파악한 후 목표 달성에 대한 계획 수립이 필요

### 1.3.4. 유지보수
- 목표 달성 후 제품이 시간이 지나면서 변할 수 있기 때문에 어떻게 유지보수해야 할지 고민해야 함
- 빠른 속도로 변화하는 AI의 특성 때문에 AI 제품 유지보수는 어려움이 있음

## 1.4. AI 엔지니어링 스택
- AI 엔지니어링은 ML 엔지니어링에서 발전해 역할 상당 부분 중복

### 1.4.1. AI의 세 가지 계층
- 최상위부터 애플리케이션 개발, 모델 개발, 인프라 세 가지 계층 존재
- 애플리케이션 개발
  - 적절한 프롬프트와 필요한 컨텍스트 제공하는 것을 포함
  - 철저한 평가 필요
  - 좋은 인터페이스 필요
- 모델 개발
  - 모델링, 학습, 파인튜닝, 추론 최적화 위한 프레임워크 포함 모델 개발하기 위한 도구 제공
  - 데이터셋 엔지니어링 포함
  - 철저한 평가 필요
- 인프라
  - 모델 서빙, 데이터와 컴퓨팅 관리, 모니터링을 위한 도구 포함
- AI 애플리케이션 개발 원칙은 동일하게 유지
  - 비즈니스 문제를 해결해야 함
    - 비즈니스 지표와 ML 지표 상호 연계 중요
  - 체계적인 실험 필요
    - 전통적인 ML 엔지니어링에서는 다양한 하이퍼파라미터로 실험
    - 파운데이션 모델에서는 서로 다른 모델, 프롬프트, 검색 알고리즘, 샘플링 변수 등으로 실험
  - 모델 빠르고 저렴하게 실행할 수 있어야 함
  - 피드백 루프를 구축해야 함

### 1.4.2. AI 엔지니어링 대 ML 엔지니어링
- 전통적인 ML의 경우 애플리케이션에 필요한 모델을 직접 학습시켜야 하지만, AI 엔지니어링에서는 다른 사람이 학습시킨 모델을 사용
- AI 엔지니어링은 전통적인 ML보다 더 크고, 많은 자원을 소비하고, 더 높은 지연 시간을 발생시켜, 효율적인 학습과 추론 최적화에 대한 압박이 더 큼
- 개방형 출력을 생성해 유연성을 제공하지만, 평가하기 더 어려움
- AI 엔지니어링은 모델 조정과 평가에 더 중점
  - 모델 조정 기법은 모델 가중치 업데이트 여부에 따라 두 가지 나뉨
    - 프롬프트 기반 기법
      - 모델 가중치를 업데이트하지 않음
      - 자유롭게 실험할 기회 제공
    - 파인튜닝
      - 모델 가중치 업데이트
      - 더 복잡하고 더 많은 데이터 필요하지만, 모델 품질, 지연 시간, 비용 크게 개선
      - 복잡한 작업에 많이 활용

#### 모델 개발
- 전통적인 ML 엔지니어링과 가장 밀접
- 모델링과 학습, 데이터셋 엔지니어링, 추론 최적화 세 가지 구성 요소가 있음

##### 모델링과 학습
- 모델 아키텍처 고안하고, 학습하고, 파인튜닝하는 과정
- ML 모델 개발하려면 전문적인 ML 지식이 필요하나, AI 애플리케이션 개발하는 데는 ML 지식이 필수가 아님

###### 학습, 사전 학습, 파인튜닝 및 사후 학습의 차이점
- 모델 가중치가 변경된다고 모두 학습은 아님
- 사전 학습, 파인튜닝, 사후 학습은 서로 다른 학습 단계이나, 학습이라는 용어로 사용될 수 있음
- 사전 학습
  - 모델을 처음부터 학습하는 것
  - 모델 가중치가 무작위로 초기화
  - 압도적으로 가장 많은 자원과 시간이 필요
- 파인튜닝
  - 이미 학습된 모델을 추가로 학습하는 것 의미
  - 모델 가중치는 이전 학습 과정에서 얻어진 값으로 진행
- 사후 학습
  - 사전 학습 이후 모델을 학습시키는 과정
  - 기술적으로 파인튜닝과 같은 과정
    - 목표에 따라 용더 구분
      - 모델 제공업체가 수행하면 사후 학습
      - 애플리케이션 개발자가 수행하면 파인튜닝
- 프롬프트 엔지니어링
  - 기술적으로 모델에 입력하는 컨텍스트를 통해 모델에게 무엇을 할지 가르치는 것

##### 데이터셋 엔지니어링
- AI 모델의 학습과 조정에 필요한 데이터를 선별하고, 생성하며, 주석을 다는 것
- 파운데이션 모델은 개방형이기 때문에, 데이터 주석이 AI 엔지니어링에서 큰 과제임
- 전통적인 ML에 비해 비정형 데이터를 더 많이 다루어 중복 제거, 토큰화, 컨텍스트 검색, 민감 정보와 유해 데이터 제거를 포함한 품질 관리에 더 중점
- **모델 자체가 대중화된 기술이 되었기 때문에 데이터가 주요 차별점**
- 필요한 데이터 양은 사용하는 조정 기법에 따라 다름
- 모델의 강점과 약점에 대한 중요한 단서를 제공하기 때문에, 데이터에 대한 전문성을 모델 검토 시 유용

##### 추론 최적화
- 모델을 더 빠르고 저렴하게 만드는 것
- 파운데이션 모델이 더 높은 추론 비용과 지연 시간을 발생시키며 규모가 커짐에 따라 중요해짐
- 자기회귀적 특징 때문에 어려움이 있음
  - 출력이 길어질수록 시간이 많이 소요되기 때문

#### 애플리케이션 개발
- 전통적인 ML 엔지니어링에서는 팀에서 자체적으로 모델을 개발했기 때문에 모델 품질이 차별화 요소
- 같은 모델을 사용하는 파운데이션 모델에서는 애플리케이션 개발 과정에서 차별화 이뤄야 함
- 평가, 프롬프트 엔지니어링, AI 인터페이스 역할로 구성

##### 평가
- 위험을 완화하고 기회를 발견하는 과정
- 구체적으로, 모델을 선택하고 진행 상황을 벤치마크하며, 애플리케이션의 배포 준비 여부를 판단하는데 쓰이고, 운영 중인 시스템의 문제점을 찾아내고 개선의 기회를 발견하는 데도 필수적
- 파운데이션 모델의 개방성과 확장된 능력, 다양한 조정 기술 때문에 ML 엔지니어링 대비 평가가 더욱 중요

##### 프롬프트 엔지니어링 및 컨텍스트 구성
- 프롬프트 엔지니어링
  - 모델 가중치를 변경하지 않고 입력만으로 AI 모델에서 원하는 동작을 이끌어 내는 것
  - 주어진 작업을 수행하는 데 필요한 컨텍스트와 도구를 모델에 제공
- 긴 컨텍스트가 필요한 복잡한 작업의 경우 자신의 기록을 추적할 수 있도록 메모리 관리 시스템 제공해야 할 수 있음 


##### AI 인터페이스
- 최종 사용자가 AI 애플리케이션과 상호작용할 수 있는 인터페이스를 만드는 것을 의미
- 이전에는 AI 모델 개발할 충분한 자원이 있는 조직만 AI 애플리케이션 개발이 가능했으나, 파운데이션 모델을 통해 독립 제품 또는 다른 사람들이 개발한 다른 제품에 탑재해 AI 애플리케이션 만들 수 있음
- 채팅 인터페이스가 가장 일반적으로 사용
- AI 인터페이스는 사용자 피드백을 수집하고 추출하는 새로운 방법도 의미

### 1.4.3. AI 엔지니어링 대 풀스택 엔지니어링
- 인터페이스에 대한 강조가 커지며 풀스택 개발에 가까워짐
- 자바스크립트 API에 대한 지원이 늘어나고 있음
- 기존 ML 엔지니어링에서는 데이터 모으고 모델 학습하는 것부터 시작했으나, 제품 먼저 만들고 가능성 있을 때에만 데이터와 모델에 투자 가능
- 기존에는 모델 개발과 제품 개발이 분리된 프로세스인 경우가 많았지만, AI 엔지니어가 제품 개발에 많이 참여하는 경향이 있음
