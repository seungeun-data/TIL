## 4.1. 평가 기준
- 평가 주도 개발 (evaluation-driven development)
  - 애플리케이션을 만들기 위해 시간과 돈을 투자하기 전에 먼저 애플리케이션을 평가할지 이해
  - AI 엔지니어링에서 개발하기 전에 평가 기준을 정의하는 것
- 일반적인 평가 기준으로, 도메인 특화 능력, 생성 능력, 지시 준수 능력, 비용과 시간으로 나눌 수 있음

### 4.1.1. 도메인 특화 능력
- 도메인 특화 능력
  - e.g. 코딩과 라틴어-영어 이해
  - 모델 구조, 크기 같은 설정이나 학습에 설정된 데이터에 따라 달라짐
- 모델이 필요한 능력을 가졌는지 평가하기 위해 공개 또는 비공개 벤치마크 활용 가능
- 코딩 관련 도메인 특화 능력
  - 정확성 평가
  - 실행 시간이나 메모리 측정해 효율성 평가
  - 코드 가독성
- 코딩 외 도메인 능력
  - 객관식 문제 같은 폐쇄형 문제로 평가
    - e.g. MMLU, AGIEval, AI2 Reasoning Challenge (ARC-C)
  - 일반적인 지표는 정확도
  - 객관식은 문제와 선택지를 조금만 다르게 표현해도 모델 성능이 달라질 수 있음
  - 폐쇄형 형태의 벤치마크가 파운데이션 모델을 평가하는 좋은 방법인지는 불분명
    - 요약, 번역, 글 작성과 같은 생성 능력 평가하기엔 적합하지 않기 때문

### 4.1.2. 생성 능력
- 자연어 생성 (Natural Language Geenration, NLG)
  - 개방형 텍스트 생성 연구
- 지표로 유창성 (fluency)과 일관성 (coherence) 포함
  - 유창성은 문장이 문법적으로 올바르고 자연스럽게 들리는지
  - 일관성은 전체 텍스트가 얼마나 잘 구조화됐는지
  - AI 평가자를 활용하거나 퍼플렉시티 사용해 평가 가능
- 각 작업은 고유 지표 존래
  - 번역의 경우, 번역문이 원문의 의미를 얼마나 잘 담아냈는지 확인하는 충실성 (faithfulness)이 있음
  - 요약의 경우, 요약문이 원본 문서에서 가장 중요한 내용을 잘 다루고 있는가에 대한 관련성 (relevance)이 있음
- 언어 모델의 생성 능력이 발전하며 지표들의 중요성이 변화
- 새로운 평가 지표 필요
  - 환각
    - 아래 유형 질의에서 환각을 일으키는 경향을 보임
      - 최소한의 지식을 요구하는 질의
      - 존재하지 않는 정보를 묻는 질의
  - 사실 관계의 일관성
    - 여러 진술, 기록, 증거, 데이터 등이 서로 모순되지 않고 동일한 사실 구조를 유지하는 상태
  - 안전성
  - 논란의 여지를 측정하는 논쟁성 (controversiality), 친근함 (friendliness), 긍정성 (positivity), 창의성 (creativity), 간결성 (concise-ness) 등

#### 사실 일관성과 비일관성
- 국소적 사실 일관성
  - 출력이 주어진 컨텍스트와 일치하면 사실 관계가 맞다고 봄
  - 작업이 제한된 영역에서 중요
- 전역적 사실 일관성
  - 출력을 공개된 지식에 기반해 평가
- 무엇이 사실인지 판단하는 게 가장 어려움
  - 어떤 출처를 신뢰하는지에 따라 달라짐
  - 모델은 웹사이트와 질의의 관련성에 지나치게 의존
- 텍스트 함의로 표현 가능
  - 전제가 주어지면, 가설이 다음 범주 중 어디 속하는지 관계 파악
    - 함의는 가설을 전제로부터 추론할 수 있음, 사실 일관성을 의미
    - 모순은 가설이 전제와 모순됨, 사실 비일관성을 의미
    - 중립은 전제는 가설을 함의하지도 모순되지도 않음, 일관성 판단할 수 없음을 의미
- 벤치마크로 TruthfulQA가 있음
  - GPT-judge라는 특수 평가자 AI와 함께 제공
- AI 평가자 활용 가능
  - GPT-3.5와 GPT-4 뛰어난 성능 보임
  - 자체 검증과 지식 강화 검증 기법 사용
    - 자체 검증
      - 모델이 서로 일치하지 않는 여러 출력을 생성하는 경우 원래 출력이 환각일 가능성이 높다고 가정
      - 평가할 응답 R이 주어지면 N개의 새로운 응답을 생성하고, R이 N개의 새로운 응답과 얼마나 일치하는지 측정
      - 많은 AI 질의가 필요해 비용이 많이 듦
    - 지식 강화 검증
  - 사실 일관성에 특화된 평가 모델 사용 가능
    - (전제, 가설) 쌍으로 입력 받아 함의, 모순, 중립 같이 미리 정의된 클래스 출력
    - e.g. DeBERTa-v3-base-mnli-fever-anli

#### 안전성
- 위험한 컨텐츠 분류
  - 욕설과 노골적인 내용을 포함한 부적절한 언어
  - 자기 파괴적 행동을 하도록 사용자를 부추기는 것과 같은 유해한 추천과 지침
  - 혐오 발언
  - 위협과 자세한 묘사를 포함한 폭력
  - 고정 관념
  - 정치적 혹은 종교적 이데올로기에 대한 편향성
- AI 평가자 사용 가능
- 사람이 작성한 텍스트의 유해성을 감지하도록 개발된 모델도 활용 가능
  - 크기가 작고, 속도가 빠르며, 비용이 적게 듦
- 일반적인 벤치마크로 RealToxicityPrompts, BOLD가 있음

### 4.1.3. 지시 수행 능력
- 모델이 주어진 지시를 얼마나 잘 따르는가
- 구조화된 출력이 필요한 애플리케이션에서 필수적
- 도메인별 능력이나 생성 능력과 혼동될 수 있어 정의하거나 측정하기 어려움

#### 지시 수행 기준
- 벤치마크마다 지시 수행 능력이 포함하는 내용이 다름
- IFEVal
  - 모델이 예상된 형식에 맞는 출력을 생성할 수 있는지에 초점
- INFOBench
  - 모델이 예상된 형식을 따를 수 있는 능력을 평가할 뿐만 아니라, 내용 제약, 언어 지침, 문체 규칙 같을 것을 따를 수 있는 능력도 평가
  - 검증을 위해 예/아니오로 답할 수 있는 질의 형태의 기준 목록 제공
  - 모델의 최종 점수는 모델이 맞힌 기준 수를 모든 지시의 총 기준 수로 나눈 것
- 벤치마크마다 평가하는 지시의 집합이 다르고 일반적으로 사용하는 많은 지시를 놓치고 있어, 벤치마크에서의 좋은 점수가 좋은 성능을 보장하진 않음 

#### 역할 연기
- 모델에게 가상 캐릭터나 페르소나 가정하도록 요청
- 목적 
  - 사용자가 상호작용할 수 있도록 캐릭터를 연기
  - 모델 출력 품질을 향상시키기 위해 프롬프트 엔지니어링 기법으로 활용
- 벤치마크로 RoleLLM, CharacterEval이 있음
  - RoleLLM은 생성된 출력과 예상 출력 간 유사도 점수와 AI 평가자 활용해 평가
  - CharacterEval은 사람 평가자를 사용해 역할 연기 관련 측면을 5점 척도로 평가하는 보상 모델 학습
- 자동 평가를 위해 휴리스틱이나 AI 평가자 사용 가능
  - 말투와 지식 두 가지 측면에서 평가 필요
  - AI 평가자 사용할 경우 역할마다 다른 프롬프트 필요

### 4.1.4. 비용과 지연 시간
- 파레토 최적화 (Pareto optimization)
  - 품질, 지연 시간, 비용 등 여러 목표 최적화
- 지연 시간
  - 첫 토큰까지 걸리는 시간, 토큰당 시간, 토큰 간 시간, 질의 당 시간 등 지표 존재
  - 모델, 프롬프트, 샘플링 변수 영향 받음
- 비용
  - 모델 API 사용 시 토큰 단위로 요금이 부과되므로 입출력 토큰 수를 줄이도록 노력
  - 자체 호스팅할 경우 보유한 장비를 최대한 활용하기 위해 장비에 맞는 가장 큰 모델 선택
  - 자체 호스팅할 경우 규모 커질수록 토큰당 비용 줄일 수 있음

## 4.2. 모델 선택
- 애플리케이션에 가장 적합한 모델이 무엇인지 중요
- 모델 조정 기법 사용하며 모델 반복적인 선택해야 함
  - (1) 달성할 수 있는 최고 성능 파악, (2) 투자 대비 최고 성능 내는 모델 선택 두 단계로 이루어짐

### 4.2.1. 모델 선택 과정
- 변경이 불가능하거나 비현실적인 하드 속성과 변경할 수 있고 기꺼이 변경할 의향이 있는 소프트 속성 구별해야 함
  - 하드 속성은 모델 제공 업체의 결정 (라이선스, 학습 데이터, 모델 크기) 또는 자체 정책 (개인 정보 보호, 제어)에 의해 결정
  - 소프트 속성은 정확도, 유해성, 사실 일관성과 같이 개선할 수 있는 속성
  - 모델과 활용 사례에 따라 다름
- 전반적인 평가 과정
  - 하드 속성이 적합하지 않은 모델은 걸러냄
  - 공개된 정보를 활용해 가장 유망한 모델 추려냄
  - 자체 평가 파이프라인으로 실험해 최적의 모델을 찾음
  - 운영 환경에서 모니터링해 실패를 감지하고 개선을 위한 피드백 수집
  - 위의 네 단계 순환적

### 4.2.2. 모델 자체 개발 대 상용 모델 구매
#### 오픈 소스, 오픈 가중치, 모델 라이선스
- 기존 오픈 소스는 사람들이 다운로드하고 사용할 수 있는 모델을 지칭
- 모델 성능이 학습 데이터에 의해 결정되므로, 학습 데이터 공개 여부에 따라, 데이터 없이 공개된 모델은 **오픈 웨이트**, 데이터와 함께 공개된 모델은 **오픈 모델**로 부름
- 현재 시점에서 오픈 소스 모델 대다수는 오픈 웨이트
- 많은 모델이 자체적인 고유 라이선스로 출시
  - 아래 내용 확인 필요
    - 상업적 사용을 허용하는가
    - 상업적 사용을 허용한다면, 어떤 제한이 있는가
    - 모델 출력을 다른 모델의 학습이나 개선에 사용되는 것을 허용하는가

#### 오픈 소스 모델과 모델 API 비교

![model api versus model hosting](../assets/images/4-2.model-api-versus-model-hosting.png)

- 추론 서비스
  - 모델 호스팅하고, 사용자 질의를 받아, 모델을 실행해 질의에 대한 응답을 생성하고, 이 응답을 사용자에게 반환하는 서비스
- 모델 API
  - 일반적으로 추론 서비스 API 지칭하나, 파인튜닝 API나 평가 API도 존재
  - 모델 제공업체, 클라우드 서비스 제공업체, 서드파티 API 제공업체를 통해 이용
- 모델 직접 호스팅할지 모델 API 사용할지 결정 시 고려 사항
  - 데이터 프라이버시
    - 조직 외부에 데이터를 전송할 수 없는 기업은 외부 호스팅 모델 API 사용 불가
  - 데이터 계보와 저작권
    - 대부분의 모델이 어떤 데이터로 학습되었는지 투명하지 않음
    - AI 보조 발명이 특허를 받을 수 없는 건 아니지만, 혁신에 대한 사람의 기여도가 특허를 받기 충분한지에 달려 있고, 저작권 있는 데이터로 학습한 모델을 써서 제품을 만들었을 떄 지적재산권을 지킬 수 있을지 불분명
    - 위와 같은 문제로 상용 모델 채택
  - 성능
    - 오픈소스 모델과 독점 모델 간 격차가 많이 좁혀졌지만, 현재 구조상 오픈소스 모델이 독점 모델을 따라잡기 어려울 것으로 보임
    - ![exclusive models versus open source models](../assets/images/4-1.exclusive-models-versus-open-source-models.png) 
  - 기능
    - 확장성 (추론 서비스가 원하는 지연 시간과 비용 유지하며 트래픽 감당하는 것), 함수 호출 (외부 도구를 사용할 수 있는 모델 능력), 출력 구조 (구조화된 출력), 출력 가드레일 (생성된 응답의 위험을 줄이는 것) 등의 기능 필요
    - 위와 같은 기능을 구현하기 어렵고 시간이 많이 걸리기 때문에, API 제공업체 이용
      - API가 제공하는 기능만 쓸 수 있다는 단점 존재
      - 파인튜닝도 제한됨
  - API 비용 대 엔지니어링 비용
    - 모델 API
      - 사용량에 따라 요금 부과해, 어느 정도 규모의 사용량인 경우 경제적이지 않음
      - SLA (서비스 수준 계약)에 의존
    - 직접 모델 호스팅
      - 상당한 시간, 기술력, 엔지니어링 노력 필요
    - 사용하고 조작하기 쉬운 모델이 좋음
      - 독점 모델은 시작하고 확장하기 좋음
      - 오픈 모델은 조작하기 쉬울 수 있음
      - 표준 API는 모델 변경하기 쉬움 
        - 가장 인기 있는 모델 API (집필 시점 오픈AI API) 모방
    - 커뮤니티 지원이 좋은 모델이 좋을 수 있음
  - 제어, 접근성, 투명성
    - API 제공업체는 항상 원하는 수준의 통제권을 주지 않음
    - 독점 모델은 안전 가드레일을 위해 검열을 강화하는 쪽으로 치우지기 쉽고, 특정 사례에서는 제약이 될 수 있음
    - 지금까지 상용 모델은 모델 변경, 버전, 로드맵에 대한 투명성 부족했음
    - 모델 제공업체가 특정 활용 사례나 산업, 국가에 대한 지원을 중단하거나 국가에서 모델 제공업체를 금지할 수도 있음
  - 온디바이스 배포
    - 인터넷 및 프라이버시 문제로 디바이스 자체에서 직접 모델 실행하고 싶은 경우 서드파티 API 사용 불가

### 4.2.3. 공개 벤치마크 탐색하기
- 모델의 다양한 능력을 평가하기 위해 수천 개의 벤치마크 존재하며, AI 활용 사례의 수에 맞춰 빠르게 증가 중
- 여러 벤치마크에서 모델 평가하는 데 도움이 되는 도구로 평가 하네스 (evaluation harness)가 있음
  - 하네스
    - 직역하면 "마구(馬具)"로, 여러 구성요소를 묶어서 제대로 작동하게 하는 장치라는 비유에서 나온 용어
    - 일반적으로 모델이나 시스템을 실행, 평가, 연결하기 위한 보조 프레임워크/도구를 의미
  - e.g. EleutherAI의 lm-evaluation-harness, 오픈AI의 evals

#### 벤치마크 선택 및 집계
- 벤치마크 결과를 집계해 순위를 매기면 리더보드가 만들어짐
  - 아래 두 가지 질문 고려해야 함
    - 어떤 벤치마크를 포함할 것인가
      - 벤치마크 선정 시 벤치마크 간 상관관계 고려해야 함
      - 집필 기간동안 많은 벤치마크가 포화 상태되거나 포화 상태에 가까워짐
    - 모델 순위를 매기기 위해서 벤치마크 결과를 어떻게 집계할 것인가
      - 평균
        - 모든 벤치마크에 같은 가중치 줌
      - 평균 승률
        - 한 모델이 다른 모델보다 더 좋은 점수를 얻는 비율을 시나리오별로 평균
- 공개 리더보드
  - 일부 벤치마크의 종합 성능 기반으로 모델 순위 매김
  - 벤치마크에서 모델 평가하려면 컴퓨팅 자원을 필요로 하기 때문에, 컴퓨팅 제약으로 소수의 벤치마크만 포함
  - 벤치마크 선정 등 리더보드 개발자들의 의사결정이 사용자에게 항상 명확하지 않음

#### 공개 벤치마크로 맞춤형 리더보드 만들기
- 애플리케이션에 필요한 핵심 능력 평가할 수 있는 벤치마크 목록 수집
  - 포화 상태가 될 수 있어 최신 벤치마크 찾아보는 게 중요
  - 벤치마크 신뢰도 평가 필요
- 모델 점수를 집계해 순위를 매길 때에는
  - 모든 벤치마크가 같은 단위나 척도를 쓰지 않음을 유의
  - 각 벤치마크가 모델에 얼마나 중요한지 생각해 점수이 가중치 선정

#### 공개 벤치마크의 데이터 오염
- 데이터 유출, 테스트 세트로 학습하기, 부정행위 등으로 불림
- 평가되는 데이터로 학습했을 때 발생
- 데이터 오염이 발생하는 방식
  - 인터넷에서 수집한 데이터로 학습되는 과정에서 실수로 공개된 벤치마크 데이터 포함
  - 평가 데이터와 학습 데이터가 동일 출처에서 오며 간접적으로 발생
  - 모델 성능 향상을 위해 고품질 벤치마크 데이터로 모델 학습
- 데이터 오염 다루기
  - 오염 감지 후 정화
    - 오염 감지 방법
      - n-gram 중복
        - 평가 샘플에 N개 토큰으로 이루어진 특정 시퀀스가 학습 데이터에도 있다면, 학습 과정에서 평가 샘플을 이미 봤을 가능성이 높음
        - 정확도 높음
        - 각 벤치마크 예시를 학습 데이터와 비교해야 해 시간과 비용이 많이 소모
        - 학습 데이터에 접근할 수 없는 경우 사용 불가
      - 퍼플렉시티
        - 모델이 평가 데이터에 대해 퍼플렉시티가 유난히 낮다면, 학습 과정에서 해당 데이터를 이미 접했을 수 있음을 의미
        - 정확도 떨어짐
        - 상대적으로 적은 자원으로도 가능

## 4.3. 평가 파이프라인 설계하기
- 자체 평가 파이프라인으로 용도에 가장 잘 맞는 모델 찾아야 함

### 4.3.1. 1단계: 시스템의 모든 구성 요소 평가하기
- 가능하다면 애플리케이션 턴별로, 작업별로 모두 평가해야 함
- 턴 기반 평가는 각 출력물의 품질 평가
- 작업 기반 평가는 시스템이 작업 완료했는지 평가
  - e.g. BIG-bench 벤치마크 모음에 twenty_questions

### 4.3.2. 2단계: 평가 가이드라인 만들기
- 가장 중요한 단계
- 애플리케이션이 해야 할 일뿐만 아니라 하면 안 되는 일 정의하는 것도 중요

#### 평가 기준 정의하기
- '좋다'는 것이 무엇을 의미하는지 정의해야 함
- '좋다'의 기준을 도출하기 위해서 실제 사용자 질의로 실험이 필요할 수 있음

#### 예시와 함께 평가 기준표 만들기
- 평가 시스템 (e.g. 이진법 (0과 1), 5점 척도 등)을 선택해야 함
- 예시와 함께 기준표를 만들어야 함

#### 평가 지표를 비즈니스 지표와 연결하기
- 애플리케이션의 지표는 해결하고자 하는 비즈니스 문제의 컨텍스트에서 고려되어야 함
- 평가 지표가 비즈니스 지표에 미치는 영향을 이해하면,
  - 계획 수립에 도움이 됨
  - 유용성 임곗값 결정하는 데도 도움

### 4.3.3. 3단계: 평가 방법과 데이터 정의하기
#### 평가 방법 선택하기
- 서로 다른 평가 기준에는 서로 다른 평가 방법 필요
- 동일한 기준에 대해 여러 평가 방법 혼합해 사용 가능
- 로그프롭은 
  - 모델이 생성된 토큰에 대해 얼마나 확신하는지 측정하는 데 사용
  - 퍼플렉시티를 통해 유창성과 사실 일관성 평가 가능
- 가능한 자동 지표를 사용하되, 필요한 경우 사람 평가에 의존
- 운영 환경에서도 평가 방법 사용함을 염두에 둬야 함

#### 평가 데이터 주석 달기
- 애플리케이션을 평가하기 위해, 가능하다면, 실제 운영 환경 데이터를 사용하고, 애플리케이션에 사용할 수 있는 자연스러운 레이블이 있으면 좋음
- 위의 데이터 및 레이블이 없다면, 사람 또는 AI 사용해 레이블 지정
- 데이터 슬라이스
  - 슬라이싱한 데이터를 하위 집합으로 나누고 각 하위 집합에 대한 시스템 성능을 개별적으로 살펴보는 것을 의미
  - 시스템을 세분하여 이해하는 목적
    - 편향 축소
    - 디버깅
    - 애플리케이션 개선 영역 발굴
    - 심슨의 역설 회피
      - 두 개의 모델 중 집계된 데이터에서 더 성능이 좋았던 모델이 모든 하위 집합에서는 또 다른 모델에 비해 성능이 떨어지는 현상을 심슨의 역설이라 함

  | 구분      | 데이터 슬라이싱 (Data Slicing) | 부트스트랩 (Bootstrap)            |
  | ------- | ----------------------- | ---------------------------- |
  | 목적      | 특정 **데이터 하위 집합 분석**     | **통계 추정의 불확실성 평가**           |
  | 방법      | 조건으로 **데이터를 나눔**        | **복원추출(resampling)** 로 샘플 생성 |
  | 중복 허용   | 없음 (보통 원 데이터 그대로)       | 있음 (같은 데이터 여러 번 뽑힘)          |
  | 주 사용 분야 | 데이터 분석, 모델 성능 분석        | 통계 추정, 신뢰구간 계산               |
  | 예       | “20대 여성 데이터만 분석”        | 데이터에서 무작위로 n개 다시 뽑아 통계량 계산   |


- 중요하게 생각하는 부분은 그에 대한 테스트셋을 만드는 것 권장
- 평가 세트에 필요한 데이터 양은 사용하는 애플리케이션과 평가 방법에 따라 다름
  - 일반적으로 평가 예시 수는 신뢰할 만큼 크되, 비용이 많이 들지 않을 정도로 작아야 함
  - 여러 부트스트랩 수행해, 매번 평가가 달라진다면, 더 큰 평가 세트가 필요함을 의미
- 평가 결과는 시스템 비교에도 사용 가능

#### 평가 파이프라인 평가하기
- 평가 파이프라인 평가 시 평가 파이프라인의 신뢰성을 향상시키고, 더 효율적으로 만드는 방법을 찾는 데 도움이 될 수 있음
- 아래 질문을 통해 평가 파이프라인 품질을 확인할 수 있음
  - 평가 파이프라인이 올바른 신호를 제공하고 있는가
  - 평가 파이프라인을 얼마나 신뢰할 수 있는가
  - 지표 간 상관관계는 어떠한가
  - 평가 파이프라인이 애플리케이션에 얼마나 많은 비용과 지연 시간을 추가하는가

#### 반복
- 요구사항과 사용자 행동이 변화하며, 평가 기준 진화할 것이고, 평가 파이프라인 반복적 개선 필요
- 일정 수준의 일관성은 유지되어야 함
- 평가 프로세스에서 변경될 수 있는 모든 변수를 기록해 실험 추적 수행
